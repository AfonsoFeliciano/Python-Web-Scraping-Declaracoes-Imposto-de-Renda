### Importando os pacotes

```python
import requests
import pandas as pd
import bs4
from bs4 import BeautifulSoup
import time
import datetime
import os
```

### Definindo os anos que serão coletados para então descobrir quais UFs possuem anos com valores

```python
anos = ['2021', '2022', '2023']

lista_ufs_anos = []

for ano in anos:

  today = datetime.date.today()
  ano_atual = str(today.year)

  if ano == ano_atual:
    url = f"https://servicos.receita.fazenda.gov.br/publico/EstatisticaIRPF/totaisDIRPF_UF.HTML"
  else:
    url = f"https://servicos.receita.fazenda.gov.br/publico/EstatisticaIRPF/totaisDIRPF_UF_{ano}.HTML"

  ufs = requests.get(url)
  print("Buscando dados no link: " + url)
  time_wait = 0.5
  print(f"Aguardando {time_wait} segundo (s).. \n")
  time.sleep(1)

  html_uf = ufs.text
  soup_uf = BeautifulSoup(html_uf, 'html.parser')

  html_final_uf = soup_uf.find_all('tr', {'class':['head even', 'head odd']})
  
  for i in soup_uf.find_all('a'):
    if len(i.text) == 2:
      lista_ufs_anos.append({"ano": ano, "uf": i.text})
```

### Exibindo a lista de anos e UFs associadas

Poderá ser observado uma lista de dicionários com todos os anos e UFs que possuem valores disponíveis

```python
lista_ufs_anos
```

### Realizando a coleta informando os Anos e UFs para retornar as Cidades

```python
lista_valores = []

for i in lista_ufs_anos:
  ano = i["ano"]
  uf = i["uf"]
  url = f"https://servicos.receita.fazenda.gov.br/publico/EstatisticaIRPF/totaisDIRPF_{uf}_{ano}.HTML"
  time_wait = 0.5
  print("Buscando dados no link: ", url)
  print(f"Aguardando {time_wait} segundo (s).. \n")
  time.sleep(time_wait)
  cidades = requests.get(url)

  html = cidades.text
  soup = BeautifulSoup(html, 'html.parser')
  html_final = soup.find_all('tr', {'class':['head even', 'head odd']})

  
  for i in (html_final):
    if "Meio\nTributação\nTipo\nResultado" not in i.get_text() and "UF\nTotal" not in i.get_text():
      lista_valores.append( (i.get_text() + uf +"\n" + ano).split('\n'))
```

### Particularidades

Como a Receita Federal Brasileira (RFB) possui diversas mudanças no decorrer dos anos, é normal que os dados coletados possuam divergências. Após algumas análises, foi observado que os dados de 2021 e 2022 não sofreram alterações, porém 2023, possui algumas colunas adicionais. Desse modo a lógica abaixo foi implementada, optando inicialmente pelo número de elementos na lista. 

Em 2021 e 2022 o número de colunas são 15, porém para 2023 o número foi aumentado. 

```python
lista_2021_2022 = []
lista_2023 = []

for i in lista_valores: 
  if len(i) == 15:
    lista_2021_2022.append(i)
  else:
    lista_2023.append(i)
```

### Definindo as colunas para cada ano e criando os DataFrames

```python
colunas_2021_2022 = ["Cidade", "Total", "Pre_Preenchida", "PGD", "APP", "Online", "Deducoes", "Simplificada", "Originais", "Retificadoras", "Sem_Imposto", "A_Pagar", "A_Restituir", "UF", "Ano"]
colunas_2023 = ["Cidade", "Total", "Pre_Preenchida", "PGD", "APP", "Online", "Deducoes", "Simplificada", "Originais", "Retificadoras", "Sem_Imposto", "A_Pagar", "A_Restituir", "Conta", "Pix", "UF", "Ano"]

df_2021_2022 = pd.DataFrame(lista_2021_2022, columns = colunas_2021_2022)
df_2023 = pd.DataFrame(lista_2023, columns = colunas_2023)

df_consolidado = df_2021_2022.merge(df_2023, how = 'outer')
```

### Convertendo dados

```python
# Criando cópia
df_final = df_consolidado.copy()

# Convertendo os dados
df_final["Total"] = df_final["Total"].astype(str)
df_final["Total"] = df_final["Total"].str.replace(".", "").astype(int)

df_final["Pre_Preenchida"] = df_final["Pre_Preenchida"].astype(str)
df_final["Pre_Preenchida"] = df_final["Pre_Preenchida"].str.replace("%", "").astype(float) / 100

df_final["PGD"] = df_final["PGD"].astype(str)
df_final["PGD"] = df_final["PGD"].str.replace("%", "").replace(",", ".").astype(float) / 100

df_final["APP"] = df_final["APP"].astype(str)
df_final["APP"] = df_final["APP"].str.replace("%", "").replace(",", ".").astype(float) / 100

df_final["Online"] = df_final["Online"].astype(str)
df_final["Online"] = df_final["Online"].str.replace("%", "").replace(",", ".").astype(float) / 100

df_final["Deducoes"] = df_final["Deducoes"].astype(str)
df_final["Deducoes"] = df_final["Deducoes"].str.replace("%", "").replace(",", ".").astype(float) / 100

df_final["Simplificada"] = df_final["Simplificada"].astype(str)
df_final["Simplificada"] = df_final["Simplificada"].str.replace("%", "").replace(",", ".").astype(float) / 100

df_final["Originais"] = df_final["Originais"].astype(str)
df_final["Originais"] = df_final["Originais"].str.replace("%", "").replace(",", ".").astype(float) / 100

df_final["Retificadoras"] = df_final["Retificadoras"].astype(str)
df_final["Retificadoras"] = df_final["Retificadoras"].str.replace("%", "").replace(",", ".").astype(float) / 100

df_final["Sem_Imposto"] = df_final["Sem_Imposto"].astype(str)
df_final["Sem_Imposto"] = df_final["Sem_Imposto"].str.replace("%", "").str.replace(",", ".").astype(float) / 100

df_final["A_Pagar"] = df_final["A_Pagar"].astype(str)
df_final["A_Pagar"] = df_final["A_Pagar"].str.replace("%", "").str.replace(",", ".").astype(float) / 100

df_final["A_Restituir"] = df_final["A_Restituir"].astype(str)
df_final["A_Restituir"] = df_final["A_Restituir"].str.replace("%", "").str.replace(",", ".").astype(float) / 100

df_final["Conta"] = df_final["Conta"].astype(str)
df_final["Conta"] = df_final["Conta"].str.replace("%", "").replace(",", ".").str.replace(",", ".").astype(float) / 100

df_final["Pix"] = df_final["Pix"].astype(str)
df_final["Pix"] = df_final["Pix"].str.replace("%", "").replace(",", ".").str.replace(",", ".").astype(float) / 100

df_final = df_final.fillna(0)


#Criando totalizadores

df_final['Total_Pre_Preenchida'] = df_final['Total'] * df_final['Pre_Preenchida']
df_final['Total_PGD'] = df_final['Total'] * df_final['PGD']
df_final['Total_APP'] = df_final['Total'] * df_final['APP']
df_final['Total_Online'] = df_final['Total'] * df_final['Online']
df_final['Total_Deducoes'] = df_final['Total'] * df_final['Deducoes']
df_final['Total_Simplificada'] = df_final['Total'] * df_final['Simplificada']
df_final['Total_Originais'] = df_final['Total'] * df_final['Originais']
df_final['Total_Retificadoras'] = df_final['Total'] * df_final['Retificadoras']
df_final['Total_Sem_Imposto'] = df_final['Total'] * df_final['Sem_Imposto']
df_final['Total_A_Pagar'] = df_final['Total'] * df_final['A_Pagar']
df_final['Total_A_Restituir'] = df_final['Total'] * df_final['A_Restituir']
df_final['Total_Conta'] = df_final['Total'] * df_final['Conta']
df_final['Total_Pix'] = df_final['Total'] * df_final['Pix']

df_final
```

```python
path_dados_consolidados = "data/dados_consolidados/"

path_dados_consolidados_exists = os.path.exists(path_dados_consolidados)
if not path_dados_consolidados_exists:

   # Create a new directory because it does not exist
   os.makedirs(path_dados_consolidados)
   print("Criando diretório: ", path_dados_consolidados)

df_final.to_excel(f"{path_dados_consolidados}dados_imposto_de_renda.xlsx", index=False)
print("Salvando arquivo excel no diretório: ", path_dados_consolidados)
```
